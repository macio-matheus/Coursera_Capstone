```python
import pandas as pd
import numpy as np
import scipy as sp
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from nxviz import CircosPlot, ArcPlot, MatrixPlot, GeoPlot
```

```python
G = nx.read_gml('graph_cervejarias_e_sorveterias.gml')
```

```python
[g for g in G.edges(data=True)]
```

```python
m = MatrixPlot(G)
m.draw()
```

```python
c = CircosPlot(G, node_color='tipo', figsize=(20, 20))
c.draw()
```

```python
a = ArcPlot(G, node_color='tipo', figsize=(20, 20))
a.draw()
```

```python
pos = nx.spring_layout(G)
betCent = nx.eigenvector_centrality_numpy(G, weight='weight')
node_color = [0 if v['tipo'] == 'pessoa' else 1 for i,v in G.nodes(data=True)]
node_size = [v * 1000 for v in betCent.values()]
plt.figure(figsize=(15,15))
nodes = G.nodes()
ed = nx.draw_networkx_edges(G, pos=pos, alpha=0.2)
no = nx.draw_networkx_nodes(G, pos=pos, with_labels=False, node_color=node_color, 
                            nodelist=nodes, cmap=cm.jet,
                             node_size=node_size)
plt.colorbar(no)
plt.axis('off')
```
