```python
import pandas as pd
import numpy as np
import scipy as sp
import networkx as nx
import matplotlib.pyplot as plt
from nxviz import CircosPlot, ArcPlot, MatrixPlot, GeoPlot
```

```python
G = nx.read_gml('graph_cervejarias_reduzido.gml')
```

```python
[g for g in G.nodes(data=True)]
```

```python
m = MatrixPlot(G, node_color='gender')
m.draw()
```

```python
c = CircosPlot(G, node_color='gender', edge_width='weight', figsize=(20, 20))
c.draw()
```

```python
a = ArcPlot(G, node_color='gender', edge_width='weight', figsize=(20, 20))
a.draw()
```

```python
pos = nx.spring_layout(G)
betCent = nx.eigenvector_centrality_numpy(G, weight='weight')
node_color = [G.degree(v, weight='weight') for v in G]
node_size = [v * 10000 for v in betCent.values()]
plt.figure(figsize=(15,15))
nodes = G.nodes()
ed = nx.draw_networkx_edges(G, pos=pos, alpha=0.2)
no = nx.draw_networkx_nodes(G, pos=pos, with_labels=False, node_color=node_color, 
                         nodelist=nodes,
                         node_size=node_size, cmap=plt.cm.jet)
plt.colorbar(no)
plt.axis('off')
```
